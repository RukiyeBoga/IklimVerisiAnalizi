```python
df['yil'] = pd.to_datetime(df['yil'], errors='coerce')
df['year'] = df['yil'].dt.year
df['month'] = df['yil'].dt.month
df['day'] = df['yil'].dt.day
df.drop(columns=['yil'], inplace=True)
```

```python
y = df[["ort_c"]].values
X = df[["IST_NO", "enlem", "boylam ", "year", "month", "day", "msn", "nem", "hPa", "saat"]].values
```

```python
# Veri setini eğitim ve test olarak ayırın
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Karar ağacı modelini oluşturun ve eğitin
model = DecisionTreeRegressor(max_depth=5, random_state=42)
model.fit(X_train, y_train)
```

```python
# Performansı değerlendirin
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f"Ortalama Kare Hata (MSE): {mse:.2f}")
```

```python
# Karar ağacını görselleştirin
plt.figure(figsize=(15, 10))
tree.plot_tree(model, feature_names=["IST_NO", "enlem", "boylam", "yil", "month", "day", "msn", "nem", "hPa", "saat"], filled=True)
plt.show()
```

```python
# Eksik sütunları doldurarak tahmin yapma
yeni_veri = [[17046, 41.1061, 42.7055, 2017,6,10, 1.1, 60, 819.1, 255.2]]  # Eksik sütunlara varsayılan değerler ekleniyor
prediction = model.predict(yeni_veri)
print(f"Tahmin edilen ortalama sıcaklık: {prediction[0]:.2f}")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Random Forest modelini oluşturun ve eğitin
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)  # n_estimators = kaç tane ağaç olacağı
rf_model.fit(X_train, y_train)

# Performansı değerlendirin
y_pred_rf = rf_model.predict(X_test)
mse_rf = mean_squared_error(y_test, y_pred_rf)
print(f"Random Forest Ortalama Kare Hata (MSE): {mse_rf:.2f}")



# Yeni veri ile tahmin yapma
yeni_veri_rf = [[17046, 41.1061, 42.7055, 2017, 6, 10, 1.1, 60, 819.1, 255.2]]
prediction_rf = rf_model.predict(yeni_veri_rf)
print(f"Random Forest ile tahmin edilen ortalama sıcaklık: {prediction_rf[0]:.2f}")

# Random Forest tahminleriyle görselleştirme (isteğe bağlı)
# DecisionTreeRegressor ile aynen yapamazsınız ama tahminleri görselleştirmek için farklı bir yaklaşım kullanabilirsiniz.
# Örneğin, tahminlerle gerçek değerleri karşılaştıran bir grafik çizebiliriz.
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_rf, color='blue')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linewidth=2)
plt.xlabel('Gerçek Değerler')
plt.ylabel('Tahmin Edilen Değerler')
plt.title('Random Forest - Gerçek ve Tahmin Edilen Değerler Karşılaştırması')
plt.show()
```
